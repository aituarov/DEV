```python
# Downloading dataset and have a look at them.
import pandas as pd
pd.set_option('display.max_rows', 100)
data = pd.read_excel('question_1_data.xlsx')
data
```

```python
# I check the AOV value
total = data['order_amount'].sum()
AOV = total / data['order_id'].count()
AOV
```

```python
# By summarising order_amount group by shop_id, we can see that shops with shop_id = 42 and shop_id=78 
# have huge value in order_amount
data.groupby(['shop_id'])['order_amount'].sum().sort_values(ascending=False)
```

```python
# If we inspect orders with shop_id = 42 and shop_id = 78, we can see abnormal huge order_amount orders
print(data.loc[((data['shop_id'] == 42) | (data['shop_id'] == 78)), :].sort_values(by='order_amount', ascending=False))
```

```python
# Calculate one pair sneakers price for each shop
df2 = data.groupby(['shop_id'])['order_amount', 'total_items'].sum()
df2['sneaker_price'] = df2['order_amount'] / df2['total_items']
print(df2.sort_values(by='sneaker_price', ascending=False))
```

```python
new_data = data.loc[(data['total_items'] < 10) & (data['order_amount'] / data['total_items'] < 400), :]
print(new_data)

AOV_new = new_data['order_amount'].sum() / new_data['order_id'].count()
print('AOV: ' + str(AOV_new))
```
